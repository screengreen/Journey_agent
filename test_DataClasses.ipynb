```python
from pydantic import BaseModel
from main.journey_llm import JourneyLLM


llm = JourneyLLM()

response = llm.invoke("Привет! Скажи одной фразой, что мы сейчас тестируем.")
print(response)
print("Ответ от модели:", response.content)
print("Провайдер:", llm.provider, "| Модель:", llm.model)
```

```python
class WeekendPlan(BaseModel):
    day: str
    activities: list[str]

llm_struct = JourneyLLM()

result = llm_struct.parse(
    output_model=WeekendPlan,
    user_prompt="Сделай простой план выходных на субботу с 3 активностями.",
    system_prompt="Ты помощник, который придумывает планы выходных.",
)

result, type(result)
```

```python
print("Day:", result.day)
print("Activities:")
for a in result.activities:
    print("-", a)
```

```python
from main.maps import YandexGeocoder, YandexRouteService


geocoder = YandexGeocoder()
route_service = YandexRouteService(geocoder=geocoder)

res = route_service.route_by_addresses(
    "Санкт-Петербург, улица Пинегина, 5",
    "Санкт-Петербург, ул. Красного Курсанта, 25Ж",
)

print("Прямое расстояние:", round(res["distance_km_straight"], 2), "км")
for mode, info in res["modes"].items():
    print(
        mode,
        "≈", round(info["distance_km"], 2), "км,",
        "время ≈", round(info["duration_min"], 1), "мин",
    )
```
